```python
import pandas as pd
import numpy as np
import pandera as pa
from nptyping import NDArray, Shape, Int64
```

```python
                   #low  high
# np.random.randint(80,  100,  size=(total,5))
total : int = 10000
column : int = 5
index_num : NDArray[Shape[str(total)],Int64] = np.arange(1,total+1)
subjects : list[str] = ['Math','Physics','Urdu','English','Islamiat']

subject_score1 : NDArray[Shape[str(total)+str(column)],Int64] = np.random.randint(80, 100,  size=(total,5))
ssubject_score1 : pd.DataFrame = pd.DataFrame(subject_score1,columns=subjects,index=index_num)
ssubject_score1

subject_score2 : NDArray[Shape[str(total)+str(column)],Int64] = np.random.randint(70, 79,  size=(total,5))
ssubject_score2 : pd.DataFrame = pd.DataFrame(subject_score2,columns=subjects,index=index_num)
ssubject_score2

subject_score3 : NDArray[Shape[str(total)+str(column)],Int64] = np.random.randint(60, 69,  size=(total,5))
ssubject_score3 : pd.DataFrame = pd.DataFrame(subject_score3,columns=subjects,index=index_num)
ssubject_score3

subject_score4 : NDArray[Shape[str(total)+str(column)],Int64] = np.random.randint(50, 59,  size=(total,5))
ssubject_score4 : pd.DataFrame = pd.DataFrame(subject_score4,columns=subjects,index=index_num)
ssubject_score4

subject_score5 : NDArray[Shape[str(total)+str(column)],Int64] = np.random.randint(40, 49,  size=(total,5))
ssubject_score5 : pd.DataFrame = pd.DataFrame(subject_score5,columns=subjects,index=index_num)
ssubject_score5

subject_score6 : NDArray[Shape[str(total)+str(column)],Int64] = np.random.randint(33, 39,  size=(total,5))
ssubject_score6 : pd.DataFrame = pd.DataFrame(subject_score6,columns=subjects,index=index_num)
ssubject_score6

subject_score7 : NDArray[Shape[str(total)+str(column)],Int64] = np.random.randint(1, 32,  size=(total,5))
ssubject_score7 : pd.DataFrame = pd.DataFrame(subject_score7,columns=subjects,index=index_num)
ssubject_score7

# display(index_num)
```

```python
df : pd.DataFrame = pd.concat([ssubject_score1,
                               ssubject_score2,
                               ssubject_score3,
                               ssubject_score4,
                               ssubject_score5,
                               ssubject_score6,
                               ssubject_score7]).reset_index(drop=True)


display(df.info())
print('============================================')
display(df.describe())
print('============================================')
display(df.head())
print('============================================')
display(df.tail())
```

```python
df.to_csv('students score',index=False)
df.to_json('std')
display(df)
```
